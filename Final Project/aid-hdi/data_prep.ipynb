```python
df = pd.read_csv('https://raw.githubusercontent.com/pmalo46/CUNY_DATA_608/master/Final%20Project/aid-hdi/person.csv')

old_columns = df.columns
new_columns=[]

for item in old_columns: 
    
    if item != 'country':
        new_columns.append('Y'+item)
        
    else:
        new_columns.append(item)
        
df.columns = new_columns
df1 = pd.wide_to_long(df, stubnames='Y', i='country', j='Year')
df1.reset_index(drop=False, inplace=True)
df1 = df1.rename(columns = {'country':'Country', 'Y':'Aid_Per_Capita'})
df1.head()
```

```python
final = final.merge(df1, on=['Country', 'Year'], how='inner')
final.head()
```

```python
final['Efficiency']=final['Change_HDI']*final['Aid_Per_Capita']
final
```

```python
region = []

for x in final['Country']:
    iso3 = Country.get_iso3_country_code_fuzzy(x)
    info = Country.get_country_info_from_iso3(iso3[0])
    reg = info.get('#region+main+name+preferred')
    region.append(reg)
    
region = pd.Series(region)
final['Region'] = region
final.head()
```
